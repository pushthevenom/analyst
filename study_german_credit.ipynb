### Влияние возраста на дефолт: Есть ли зависимость между возрастом клиента и вероятностью дефолта по кредиту?

Чтобы определить косвенные признаки дефолта, включая влияние возраста, мы разобьем клиентов на возрастные группы. Затем агрегируем данные для вычисления количества и среднего процента просрочек по дефолту в каждой группе.   
Для наглядности построим график.  

```python
# возрастные группы
# значение 8 можно изменить на любое другое,таким образом мы можем смотреть на группы в разных масштабах
df['age_group'] = pd.qcut(df['age'],8)
```

```python
df.groupby('age_group')['default'].agg(['mean','count'])
```

```python
# график для общей наглядности
df.groupby('age_group')['default'].mean().plot(ylim=0,rot=90,grid=True)
```

- Наблюдается тенденция к увеличению процента просрочек по кредиту в более молодых возрастных группах, особенно среди клиентов 18-24 лет. За ними следуют клиенты 24-27 и 36-42 лет.   
- Самый низкий процент просрочек отмечен в возрастной группе 30-33 года.

```python
df.groupby('sex')['default'].agg(['mean','count'])
```

- В нашей выборке присутствует гендерный дисбаланс: мужчин в два раза больше, чем женщин. При этом процент просрочек по кредиту выше у женщин, чем у мужчин.

### Влияние цели кредита на дефолт: Есть ли различия в показателях дефолта для кредитов, взятых на разные цели?

Показатель Purpose в таблице представляет собой текстовое описание цели, для которой клиент берет кредит. Он классифицирует кредиты по их назначению и включает следующие категории:  

car: Кредит на покупку автомобиля.  
furniture/equipment: Кредит на покупку мебели или оборудования для дома.  
radio/TV: Кредит на покупку радио-, телевизионной или другой подобной техники.  
domestic appliances: Кредит на покупку бытовой техники (например, холодильник, стиральная машина).  
repairs: Кредит на проведение ремонтных работ (например, ремонт квартиры или дома).  
education: Кредит на оплату обучения (например, образование в университете, профессиональные курсы).  
business: Кредит на развитие или открытие бизнеса.  
vacation/others: Кредит на отпуск, путешествия или другие личные нужды, не включенные в другие категории.  

```python
df.groupby('purpose')['default'].value_counts()
```

- Большинство кредитов выдается на покупку автомобилей и радио/ТВ.

```python
df.groupby('purpose')['sex'].value_counts()
```

- по большинству целей кредита, автомобиль, радио/ТВ, мебель/оборудование и бизнес, мужчины берут кредиты чаще, чем женщины.   
- Кредиты на образование и бытовую технику берут примерно поровну, но общая выборка по этим категориям мала.
- В категориях “ремонт” и “отпуск/другое” наблюдается крайне малое количество кредитов.”
